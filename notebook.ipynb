```python
import pandas as pd 
import numpy as np
import seaborn as sb
```

```python
data = pd.read_csv('Meteorite_Landings.csv', sep=',')
data = data.head(60)
data = data['mass (g)']
mean = np.mean(data)
median = np.nanmedian(data)

print('media:',mean)
print('mediana:', median)
print('moda:',data.value_counts().idxmax())
sb.histplot(data, bins=10, kde=True)
```
